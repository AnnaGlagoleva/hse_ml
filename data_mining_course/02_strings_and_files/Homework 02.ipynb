1 (1 балл)
На вход вам приходит строка, в которой буква h встречается минимум два раза. Выведите эту строку без первого и последнего вхождение буквы h, а также без всех символов, находящихся между ними.

Пример входа: 'habrahabr'

Пример выхода: 'abr' 


Пример входа: 'AAAAAhBBBBBhCCCCChDDDDDD'

Пример выхода: 'AAAAADDDDDD' 


Пример входа: 'hh'

Пример выхода: ''

```python
h_string = 'AAAAAhBBBBBhCCCCChDDDDDD'
left_h_index = h_string.find('h')
right_h_index = h_string.rfind('h')

h_string_result = h_string[:left_h_index] + h_string[right_h_index+1:]
h_string_result
```

```python
poem = """5 little piges.
This little pig went to market,
This little pig stayed at home,
This little pig had roast beef,
And this little pig had none,
This little pig said, "Wee, wee, wee!
I can't find my way home."""

poem_cleaned = poem.replace('.', '').replace('"', '').replace('!', '').replace(',', '')
print('poem_cleaned:')
print(poem_cleaned)
print()

poem_cleaned_lower = poem_cleaned.lower()
print('poem_cleaned_lower:')
print(poem_cleaned_lower)
print()

poem_cleaned_lower_split = poem_cleaned_lower.split()
print('poem_cleaned_lower_split:')
print(poem_cleaned_lower_split)

len(set(poem_cleaned_lower_split))
```

3 (3 балла)
Скачайте файл с распространенными паролями: https://raw.githubusercontent.com/danielmiessler/SecLists/master/Passwords/Common-Credentials/10-million-password-list-top-500.txt

Откройте файл в Питоне с помощью with. Проитерируйтесь по строкам и проверьте каждый пароль на надежность. Будем считать пароль надежным, если в нем минимум 6 символов разного регистра, есть цифры, буквы и другие символы. Выведете долю надежных паролей среди всех в файле.

В решении могут помочь методы .isupper() , .isdigit() , .isalpha()

```python
with open('10-million-password-list-top-500.txt') as passords_fp:
    passwords_count = 0
    strong_passwords_count = 0
    for line in passords_fp:
        has_digit = False
        has_upper = False
        has_lower = False
        passwords_count += 1
        print(line[:-1], end='')
        
        for char in line[:-1]:
            if char.isdigit():
                has_digit = True
            if char.isalpha():
                if char.isupper():
                    has_upper = True
                else:
                    has_lower = True
                    
        if has_digit and has_upper and has_lower:
            strong_passwords_count += 1
            print(line[:-1], end='')
            print(' Strong')
        else:
            print(' Not strong')
            
print('\nStrong passwords ratio:', strong_passwords_count / passwords_count)
```
