```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms
import community
from networkx.algorithms import community as community_methods
```

```python
#Давайте посмотрим на наш привычный клуб каратистов

G = nx.karate_club_graph()
```

```python
communities = community.best_partition(G)

communities_generator= community_methods.label_propagation_communities(G)


communities = [communities.get(node) for node in G.nodes()]
node_color = list(communities)

nx.draw_spring(G, node_color=node_color, node_size=20, width=0.1)
plt.show()
```
